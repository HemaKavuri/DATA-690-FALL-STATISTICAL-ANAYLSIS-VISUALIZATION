**You must pick a few countries of your interest and compare them to see how they differ in the indicator**

```python
country_dict={'Malaysia','Brazil','Afghanistan','United Kingdom','Zimbabwe'}
countries=world_development[world_development['Country Name'].isin(country_dict)]
countries
```

**You must pick one indicator out of the four: population, life expectancy, GDP per capita, or suiside rate**

```python
sucide_indicator=countries.loc[countries['indicator']=='SH.STA.SUIC.P5']
sucide_indicator.head()
```

```python
sucide_indicator['Year'].max()
```

**Compare one year (pick the latest year) using bar chart, pie chart, and boxplot**

```python
year=sucide_indicator.loc[sucide_indicator['Year']==2019]
year.head()
```

**Bar Plot**

```python
fig=px.bar(year,x='Country Name',y='value', color='Country Name',title='Sucide Rate in 2019')
fig.show()
```

**Pie Chart**

```python
fig = px.pie(year, values='value', names='Country Name', title='Sucide Rate in 2019')
fig.show()
```

**Box Plot**

```python
fig=px.box(year,x='Country Name',y='value', color='Country Name',title='Sucide Rate in 2019')
fig.show()
```

**Compare all years using line chart**





```python
fig=px.line(sucide_indicator,x='Year',y='value', color='Country Name',title='Sucide Rate in all years')
fig.show()
```

```python
sucide_indicator1=world_development.loc[world_development['indicator']=='SH.STA.SUIC.P5']
sucide_indicator1.head()
```

```python
year=sucide_indicator1.loc[sucide_indicator1['Year']==2019]
year.head()
```

```python
regions_average=year.groupby('Region')['value'].mean()
regions_average
```

```python
fig=px.bar(regions_average,x=regions_average.index,y='value', color=regions_average.index,title='Sucide Rate in 2019')
fig.show()
```

```python
fig = px.pie(regions_average, values='value', names=regions_average.index, title='Sucide Rate in 2019')
fig.show()
```

**Compare all years using Line Chart**

```python
regions_average_all=sucide_indicator1.groupby(['Region','Year'])['value'].mean()
regions_average_all=regions_average_all.reset_index()
```

```python
fig=px.line(regions_average_all,x='Year',y='value',color='Region',title='Sucide Rate in all years')
fig.show()
```
